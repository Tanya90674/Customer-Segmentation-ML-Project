```python
features = ["Age", "Income", "Amount_Spent", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency"]
```

```python
X=df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_Scaled = scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2,10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_Scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10),wcss,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6)
df["Clusters"] = kmeans.fit_predict(X_Scaled)
df["Clusters"].value_counts().sort_index()
```

```python
df.head()
```

```python
cluster_summary = df.groupby("Clusters")[features].mean()
```

```python
cluster_summary
```

```python
cluster_summary.round(2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_Scaled)
df["PCA1"], df["PCA2"] = pca_data[:,0], pca_data[:,1]
```

```python
pca_data
```

```python
sns.scatterplot(x="PCA1",y="PCA2",hue="Clusters",data=df,palette="Set1" )
plt.title("Customer Segmentation based on PCA")
plt.show()
```

```python
import joblib
joblib.dump(kmeans,"kmeans_model.pkl")
joblib.dump(scaler,"scaler.pkl")
```
